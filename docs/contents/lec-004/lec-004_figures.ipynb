## Figure 1. Euler product formula가 성립하는 복소평면 영역 $\operatorname{Re}(s) > 1$

```python
a     = 0.6
sigma = 2.5
t     = 1.5

x_min, x_max = -1.2, 3.5
y_min, y_max = -2.2, 2.2

fig, ax = plt.subplots(figsize=(3, 3))

ax.set_xlim(x_min, x_max)
ax.set_ylim(y_min, y_max)

# Green shaded half-plane Re(s) > 1
ax.fill_betweenx([y_min, y_max], 1, x_max,
                 color="lightgreen", alpha=0.4, zorder=1)

ax.spines["left"].set_position("zero")
ax.spines["bottom"].set_position("zero")
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.plot(x_max, 0,     ">k", ms=MS_ARROW, clip_on=False)
ax.plot(0,     y_max, "^k", ms=MS_ARROW, clip_on=False)

# Axis labels
ax.text(x_max + 0.1, 0, "Real", va="center", fontsize=FS_LABEL)
ax.text(0.08, y_max - 0.1, "Imaginary", ha="left",   fontsize=FS_LABEL)

# Origin label
ax.text(-0.1, -0.2, "0", ha="right", va="top", fontsize=FS_TICK)

# s = 1: open circle (pole) + dashed vertical line
ax.plot([1, 1], [y_min, y_max], "k--", lw=LW_TICK, zorder=2)
ax.plot(1, 0, "o", color="white", ms=MS_POLE,
        markeredgecolor="black", markeredgewidth=1.0, zorder=4)
ax.text(0.9, -0.2, "1", ha="center", va="top", fontsize=FS_TICK)

ax.set_xticks([])
ax.set_yticks([])

plt.tight_layout()
os.makedirs("images", exist_ok=True)
plt.savefig(os.path.join("images", "fig-004-001.png"), bbox_inches="tight")
plt.show()
```
